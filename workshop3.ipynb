```python
# Importing necessary libraries
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

# **Task 3.1: Fare Revenue Proportion by Class**
# Calculate the total Fare paid across all passengers
total_fare = titanic['Fare'].sum()

# Calculate total Fare for each class
first_class_fare = titanic[titanic['Pclass'] == 1]['Fare'].sum()
second_class_fare = titanic[titanic['Pclass'] == 2]['Fare'].sum()
third_class_fare = titanic[titanic['Pclass'] == 3]['Fare'].sum()

# Combine the fare totals from each class
fare_totals = [first_class_fare, second_class_fare, third_class_fare]

# Calculate the proportion of fare revenue by class
fare_proportions = [fare / total_fare * 100 for fare in fare_totals]

# Display results
print("Fare Revenue Proportion by Class:")
print(f"First Class: {fare_proportions[0]:.2f}%")
print(f"Second Class: {fare_proportions[1]:.2f}%")
print(f"Third Class: {fare_proportions[2]:.2f}%")

# **Task 3.2: Age Group Proportion**
# Create a new column categorizing age groups
def categorize_age(age):
    if age < 18:
        return "child"
    elif 18 <= age < 65:
        return "adult"
    else:
        return "senior"

# Handle missing Age values by filling with the median age
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())

# Add age_group column
titanic['age_group'] = titanic['Age'].apply(categorize_age)

# Calculate the total number of passengers
total_passengers = len(titanic)

# Count passengers in each age group
age_group_counts = titanic['age_group'].value_counts()

# Calculate the proportion of passengers in each age group
age_group_proportions = (age_group_counts / total_passengers) * 100

# Display results
print("\nPassenger Proportion by Age Group:")
for group, proportion in age_group_proportions.items():
    print(f"{group.capitalize()}: {proportion:.2f}%")
```
